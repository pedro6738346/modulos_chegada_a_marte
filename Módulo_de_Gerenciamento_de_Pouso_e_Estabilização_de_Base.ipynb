```python
fila_pouso = [
    {"nome": "Energia",
     "prioridade": 5,
     "criticidade": "Alta",
     "combustivel": 80,
     "massa": 15,
     "chegada": "06:00",
     "sensores": "OK",
     "area_de_pouso": "OK",
     "atmosfera": "OK"},

    {"nome": "Plantações",
     "prioridade": 5,
     "criticidade": "Alta",
     "combustivel": 90,
     "massa": 12,
     "chegada": "07:00",
     "sensores": "OK",
     "area_de_pouso": "OK",
     "atmosfera": "ERRO"},

    {"nome": "Laboratório",
     "prioridade": 5,
     "criticidade": "Alta",
     "combustivel": 90,
     "massa": 10,
     "chegada": "08:00",
     "sensores": "ERRO",
     "area_de_pouso": "OK",
     "atmosfera": "OK"},

    {"nome": "Medicina",
     "prioridade": 4,
     "criticidade": "Média",
     "combustivel": 85,
     "massa": 8,
     "chegada": "09:00",
     "sensores": "OK",
     "area_de_pouso": "OK",
     "atmosfera": "OK"},

    {"nome": "Habitação",
     "prioridade": 3,
     "criticidade": "Média",
     "combustivel": 15,
     "massa": 18,
     "chegada": "10:00",
     "sensores": "OK",
     "area_de_pouso": "OK",
     "atmosfera": "OK"},

    {"nome": "Desenv. Tecnológico",
     "prioridade": 2,
     "criticidade": "Média",
     "combustivel": 95,
     "massa": 6,
     "chegada": "11:00",
     "sensores": "OK",
     "area_de_pouso": "ERRO",
     "atmosfera": "OK"}
]

modulos_pousados = []
pilhas_alertas = []

def adicionar_alerta(fila_pouso):
    for modulo in fila_pouso:
        if (modulo["combustivel"] > 20 and
                modulo["sensores"] == "OK" and
                modulo["area_de_pouso"] == "OK" and
                (modulo["atmosfera"] == "OK" or
                 modulo["criticidade"] == "Alta")):
            continue
        else:
            pilhas_alertas.append(modulo)
    return pilhas_alertas

def resolver_alerta(pilhas_alertas):
    if len(pilhas_alertas) > 0:
        resultado = pilhas_alertas.pop()
        return resultado
    else:
        return "Sem alertas"

def buscar_menor_combustivel(fila_pouso):
    menor_combustivel = None
    for modulo in fila_pouso:
        if menor_combustivel is None or modulo["combustivel"] < menor_combustivel["combustivel"]:
            menor_combustivel = modulo
    return menor_combustivel["nome"]

def buscar_maior_prioridade(fila_pouso):
    maior_prioridade = None
    for modulo in fila_pouso:
        if maior_prioridade is None or modulo["prioridade"] > maior_prioridade["prioridade"]:
            maior_prioridade = modulo
    return maior_prioridade["nome"]

def buscar_por_criticidade(fila_pouso, nivel):
    resultado = []
    for modulo in fila_pouso:
        if modulo["criticidade"] == nivel:
            resultado.append(modulo)
    return resultado

def ordenar_por_prioridade(fila_pouso):
    n = len(fila_pouso)
    for i in range(n):
        for j in range(0, n-i-1):
            if fila_pouso[j]["prioridade"] < fila_pouso[j+1]["prioridade"]:
                fila_pouso[j], fila_pouso[j+1] = fila_pouso[j+1], fila_pouso[j]
    return fila_pouso

def simular_pouso(fila_pouso):
    while fila_pouso:
        modulo = fila_pouso.pop(0)
        if (modulo["combustivel"] > 20 and
                modulo["sensores"] == "OK" and
                modulo["area_de_pouso"] == "OK" and
                (modulo["atmosfera"] == "OK" or
                 modulo["criticidade"] == "Alta")):
            modulos_pousados.append(modulo)
        else:
            pilhas_alertas.append(modulo)


print("=== Ordenando fila por prioridade ===")
ordenar_por_prioridade(fila_pouso)
for modulo in fila_pouso:
    print(f"{modulo['nome']} - Prioridade: {modulo['prioridade']}")

print("\n=== Simulando pousos ===")
simular_pouso(fila_pouso)


print("\n=== Módulos pousados ===")
for modulo in modulos_pousados:
    print(f"✅ {modulo['nome']}")


print("\n=== Módulos em alerta ===")
for modulo in pilhas_alertas:
    print(f"❌ {modulo['nome']}")


print("\n=== Buscas ===")
print(f"Menor combustível: {buscar_menor_combustivel(modulos_pousados)}")
print(f"Maior prioridade: {buscar_maior_prioridade(modulos_pousados)}")
print(f"Criticidade Alta: {[m['nome'] for m in buscar_por_criticidade(modulos_pousados, 'Alta')]}")


print("\n=== Resolvendo alerta ===")
print(f"Alerta resolvido: {resolver_alerta(pilhas_alertas)}")
```
